feature engineering

```python
df['Hour'] = (df['Time'] // 3600) % 24
df['Day'] = (df['Time'] // (3600*24)) % 7
df['Period'] = pd.cut(df['Hour'], bins=[0,6,12,18,24], labels=['Night','Morning','Afternoon','Evening'], right=False)
df
```

```python
df['Amount_log'] = np.log1p(df['Amount'])  # логарифм для сглаживания выбросов
df
```

```python
#Статистические признаки для скользящего окна
df['Amount_diff'] = df['Amount'].diff().fillna(0)
df['Amount_pct_change'] = df['Amount'].pct_change().fillna(0)
df
```

```python
#взаимодействия признаков
df['V_sum'] = df[[f'V{i}' for i in range(1,29)]].sum(axis=1)
df['V_mean'] = df[[f'V{i}' for i in range(1,29)]].mean(axis=1)
df['V_std'] = df[[f'V{i}' for i in range(1,29)]].std(axis=1)
df
```

```python
#категориальные признаки
df = pd.get_dummies(df, columns=['Period'], drop_first=True)
```

```python
print(df[['Hour','Day','Amount_log','Amount_diff','Amount_pct_change','V_sum','V_mean','V_std']].head())
```

```python
df['V_sum_diff'] = df['V_sum'].diff().fillna(0)
df['V_mean_diff'] = df['V_mean'].diff().fillna(0)
df['V_std_diff'] = df['V_std'].diff().fillna(0)
```

```python
df['Amount_rolling_mean_3'] = df['Amount'].rolling(3).mean().fillna(df['Amount'])
df['Amount_rolling_std_3'] = df['Amount'].rolling(3).std().fillna(0)
```

```python
df['Amount_per_V_sum'] = df['Amount'] / (df['V_sum'] + 1e-5)
df['Amount_per_V_mean'] = df['Amount'] / (df['V_mean'] + 1e-5)
```

```python
from sklearn.metrics import roc_auc_score, classification_report, confusion_matrix

def compare_models(models, X_test, y_test, thresholds=None):
    for name, model in models.items():
        print(f"\n=== {name} ===")
        y_prob = model.predict_proba(X_test)[:, 1]
        
        # Применяем порог, если указан
        threshold = thresholds[name] if thresholds and name in thresholds else 0.5
        y_pred = (y_prob >= threshold).astype(int)
        
        print("ROC AUC:", roc_auc_score(y_test, y_prob))
        print(classification_report(y_test, y_pred))
        print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))

# Использование:
models_dict = {'model1': model1, 'model2': model2}
thresholds_dict = {'model1': 0.5, 'model2': 0.9}

compare_models(models_dict, X_test, y_test, thresholds_dict)
```

```python
import pandas as pd
from sklearn.metrics import roc_auc_score, precision_score, recall_score, f1_score, accuracy_score, confusion_matrix

def metrics_table(models, X_test, y_test, thresholds=None):
    rows = []
    for name, model in models.items():
        y_prob = model.predict_proba(X_test)[:, 1]
        threshold = thresholds[name] if thresholds and name in thresholds else 0.5
        y_pred = (y_prob >= threshold).astype(int)
        
        tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()
        row = {
            "Model": name,
            "ROC AUC": roc_auc_score(y_test, y_prob),
            "Accuracy": accuracy_score(y_test, y_pred),
            "Precision (fraud)": precision_score(y_test, y_pred),
            "Recall (fraud)": recall_score(y_test, y_pred),
            "F1 (fraud)": f1_score(y_test, y_pred),
            "TP": tp,
            "FP": fp,
            "TN": tn,
            "FN": fn
        }
        rows.append(row)
    
    return pd.DataFrame(rows)

models_dict = {"model1": model1, "model2": model2}
thresholds_dict = {"model1": 0.5, "model2": 0.9}

df_comparison = metrics_table(models_dict, X_test, y_test, thresholds_dict)
print(df_comparison)
```

```python
from sklearn.metrics import roc_auc_score, classification_report, confusion_matrix, f1_score

def evaluate_model(model, X, y, threshold=0.5):
    y_prob = model.predict_proba(X)[:, 1]
    
    y_pred = (y_prob >= threshold).astype(int)
    
    roc_auc = roc_auc_score(y, y_prob)
    f1 = f1_score(y, y_pred)
    report = classification_report(y, y_pred, digits=4)
    cm = confusion_matrix(y, y_pred)
    
    print(f"=== Evaluation at threshold {threshold} ===")
    print("ROC AUC:", round(roc_auc, 4))
    print("F1-score:", round(f1, 4))
    print("Classification Report:\n", report)
    print("Confusion Matrix:\n", cm)
    print("\n")
    
    return {
        "ROC_AUC": roc_auc,
        "F1": f1,
        "ConfusionMatrix": cm
    }
```

```python
X_new = X_test  # признаки
y_new = y_test  # целевая переменная
```

```python
metrics_model1 = evaluate_model(model1, X_new, y_new, threshold=0.9)
metrics_model2 = evaluate_model(model2, X_new, y_new, threshold=0.9)
```

```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

models = {'model1': model1, 'model2': model2}
plt.figure(figsize=(8,6))

for name, mdl in models.items():
    y_prob = mdl.predict_proba(X_new)[:, 1]
    fpr, tpr, _ = roc_curve(y_new, y_prob)
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, label=f'{name} (AUC = {roc_auc:.3f})')

plt.plot([0,1], [0,1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve Comparison')
plt.legend()
plt.show()
```
